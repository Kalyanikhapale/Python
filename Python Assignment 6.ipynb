```python
'''Q.5 Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap'''
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(42)

# Generate synthetic correlated features
n_samples = 100
x1 = np.random.normal(0, 1, n_samples)
x2 = 2 * x1 + np.random.normal(0, 0.5, n_samples)  # strongly correlated with x1
x3 = np.random.normal(0, 1, n_samples)             # not correlated

# Create a DataFrame
df = pd.DataFrame({
    'Feature 1': x1,
    'Feature 2': x2,
    'Feature 3': x3
})

# Compute correlation matrix
corr_matrix = df.corr()

# Visualize correlation matrix using a heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
#1. Violin Plot  Student Grade Distribution
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate Student Grades Data
np.random.seed(15)
data_grades = {
    'Grade': np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),
    'Score': np.random.randint(50, 100, 200)
}
df_grades = pd.DataFrame(data_grades)

# Create Violin Plot
plt.figure(figsize=(8, 5))
sns.violinplot(x='Grade', y='Score', data=df_grades, palette='pastel')

# Labels and Title
plt.title('Distribution of Student Scores by Grade')
plt.xlabel('Grade')
plt.ylabel('Score')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate sales data
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

# Pivot the data to create a matrix of sales (Days as rows, Months as columns)
sales_pivot = df.pivot_table(values='Sales', index='Day', columns='Month', aggfunc='mean')

# Create the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(sales_pivot, annot=True, fmt=".0f", cmap='coolwarm', linewidths=0.5, linecolor='gray')

# Labels and title
plt.title('Sales Variation by Day and Month')
plt.xlabel('Month')
plt.ylabel('Day')
plt.tight_layout()
plt.show()
```

```python
'''Q.4 Using the given x and y data, generate a 3D surface plot to visualize the function  z= sin (x^2 + y^2)
x = np.linspace(-5, 5, 100)
 y = np.linspace(-5, 5, 100)
 x, y = np.meshgrid(x, y)
 z = np.sin(np.sqrt(x**2 + y**2))
 data = {
 'X': x.flatten(),
 'Y': y.flatten(),
 'Z': z.flatten()
 }

 df = pd.DataFrame(data)'''
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate meshgrid data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Optional: Flatten and store in DataFrame
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create 3D surface plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surf = ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')

# Add color bar and labels
fig.colorbar(surf, shrink=0.5, aspect=10)
ax.set_title('3D Surface Plot of z = sin((x + y))')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate synthetic data
np.random.seed(25)
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)

# Create bubble chart
plt.figure(figsize=(10, 6))
plt.scatter(df['GDP'], df['Population'],
            s=df['Population']*5,  # scale bubble size
            alpha=0.6, c='skyblue', edgecolors='black')

# Add country labels
for i in range(len(df)):
    plt.text(df['GDP'][i] + 10, df['Population'][i], df['Country'][i], fontsize=10)

# Labels and title
plt.xlabel('GDP (in billions)')
plt.ylabel('Population (in millions)')
plt.title('Bubble Chart: Population vs GDP by Country')
plt.grid(True)
plt.show()
```

```python
'''Q.1 Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x'''
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Enable Bokeh output in notebooks (remove if running as standalone script)
output_notebook()

# Data
x = np.linspace(0, 10, 500)
y = np.sin(x)

# Create Bokeh figure
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)', width=700, height=400)
p.line(x, y, line_width=2, color='navy', legend_label="sin(x)")

# Customize
p.legend.location = "top_left"
p.grid.grid_line_alpha = 0.3

# Show the plot
show(p)
```

```python
'''Q.2 Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.'''
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource
import numpy as np
import pandas as pd

# For notebook display (optional)
output_notebook()

# Generate random data
np.random.seed(42)
n = 100
data = {
    'x': np.random.uniform(0, 10, n),
    'y': np.random.uniform(0, 10, n),
    'sizes': np.random.randint(5, 20, n),
    'colors': np.random.choice(['red', 'green', 'blue', 'orange', 'purple'], n)
}
df = pd.DataFrame(data)

# Convert to ColumnDataSource for Bokeh
source = ColumnDataSource(df)

# Create scatter plot
p = figure(title="Bokeh Scatter Plot with Variable Sizes and Colors",
           x_axis_label='X Axis', y_axis_label='Y Axis', width=700, height=400)

p.scatter(x='x', y='y', size='sizes', color='colors', source=source, alpha=0.6)

# Show the plot
show(p)
```

```python
'''Q.3  Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
 fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
 counts = [20, 25, 30, 35]'''
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource
from bokeh.transform import dodge
from bokeh.palettes import Category10

# Display in notebook (optional)
output_notebook()

# Data
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts, color=Category10[4]))

# Create figure
p = figure(x_range=fruits, height=400, title="Fruit Counts",
           toolbar_location=None, tools="")

# Add vertical bars
p.vbar(x='fruits', top='counts', width=0.6, color='color', legend_field='fruits', source=source)

# Customize plot
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.yaxis.axis_label = "Count"
p.xaxis.axis_label = "Fruit"
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

# Show plot
show(p)
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Display plot in notebook (optional)
output_notebook()

# Generate data and histogram bins
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Create Bokeh figure
p = figure(title='Histogram of Random Data',
           x_axis_label='Value',
           y_axis_label='Frequency',
           width=700, height=400)

# Draw the histogram using quad glyphs
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
       fill_color='navy', line_color='white', alpha=0.7)

# Show the plot
show(p)
```

```python
'''Q.5  Create a Bokeh heatmap using the provided dataset.
 data_heatmap = np.random.rand(10, 10)
 x = np.linspace(0, 1, 10)
 y = np.linspace(0, 1, 10)
 xx, yy = np.meshgrid(x, y)'''
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, LinearColorMapper, ColorBar
from bokeh.transform import transform
import numpy as np

# Enable output in notebook (optional)
output_notebook()

# Generate heatmap data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Flatten the data for Bokeh
source = ColumnDataSource(data=dict(
    x=xx.flatten(),
    y=yy.flatten(),
    value=data_heatmap.flatten()
))

# Create a color mapper
color_mapper = LinearColorMapper(palette='Viridis256', low=data_heatmap.min(), high=data_heatmap.max())

# Create figure
p = figure(title="Bokeh Heatmap",
           x_range=(0, 1), y_range=(0, 1),
           tools="", toolbar_location=None,
           width=500, height=500)

# Draw heatmap squares
p.rect(x='x', y='y', width=0.1, height=0.1,
       source=source,
       fill_color=transform('value', color_mapper),
       line_color=None)

# Add color bar
color_bar = ColorBar(color_mapper=color_mapper, label_standoff=8, location=(0, 0))
p.add_layout(color_bar, 'right')

# Show the heatmap
show(p)
```
