1. Difference between np.array, np.asarray, and np.asanyarray

These three NumPy functions convert inputs into arrays, but they have different behaviors regarding copying and subclass handling.

| Function          | Copy Behavior                                 | Accepts Subclasses (like `np.matrix`) | Typical Use                              |
| ----------------- | --------------------------------------------- | ------------------------------------- | ---------------------------------------- |
| `np.array()`      | **Always makes a copy** (unless `copy=False`) | ❌ Converts subclasses to `ndarray`    | Ensure a **new array** is made           |
| `np.asarray()`    | Avoids copy if input is already an array      | ❌ Converts subclasses to `ndarray`    | Avoid unnecessary copies                 |
| `np.asanyarray()` | Avoids copy and **preserves subclasses**      | ✅ Keeps `np.matrix`, etc.             | Preserve subclass types like `np.matrix` |


```python
import numpy as np

a = np.array([1, 2, 3])
b = np.asarray(a)         # No copy made, shares memory with a
c = np.asanyarray(a)      # Same as b, also shares memory

m = np.matrix([1, 2, 3])
print(type(np.asarray(m)))     # <class 'numpy.ndarray'>
print(type(np.asanyarray(m)))  # <class 'numpy.matrix'>
```

 2. Difference between Deep Copy and Shallow Copy

Shallow Copy:

-Creates a new object, but does not copy the nested objects.

-Changes in the inner elements affect the original.

Deep Copy:

-Creates a completely independent clone, including all nested objects.

-Changes do not affect the original object.

```python
import copy

original = [[1, 2], [3, 4]]

shallow = copy.copy(original)
deep = copy.deepcopy(original)

# Modify inner list
shallow[0][0] = 99
deep[1][1] = 88

print("Original:", original)   # Affected by shallow
print("Shallow:", shallow)
print("Deep:", deep)           # Unaffected
```

| Feature               | Shallow Copy                  | Deep Copy                  |
| --------------------- | ----------------------------- | -------------------------- |
| Copies outer object?  | ✅ Yes                         | ✅ Yes                      |
| Copies inner objects? | ❌ No                          | ✅ Yes                      |
| Uses in NumPy         | `view()` creates shallow copy | `copy()` creates deep copy |

```python
'''Q.14 Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and
Perform the following tasks using Matplotlib:
 right y-axis for 'Humidity').
 b) Label the x-axis as 'Date'.
 a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and
c) Set the title of the plot as 'Temperature and Humidity Over Time'.'''
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create a time-series DataFrame
date_range = pd.date_range(start='2024-01-01', periods=30, freq='D')

# Generate sample data
temperature = np.random.uniform(15, 35, size=30)   # Degrees Celsius
humidity = np.random.uniform(40, 90, size=30)      # Percentage

# Create DataFrame
df = pd.DataFrame({
    'Date': date_range,
    'Temperature': temperature,
    'Humidity': humidity
})

# Step 2: Plotting
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot Temperature on left y-axis
ax1.plot(df['Date'], df['Temperature'], color='red', label='Temperature (°C)')
ax1.set_ylabel('Temperature (°C)', color='red')
ax1.tick_params(axis='y', labelcolor='red')

# Step 3: Create a second y-axis for Humidity
ax2 = ax1.twinx()
ax2.plot(df['Date'], df['Humidity'], color='blue', label='Humidity (%)')
ax2.set_ylabel('Humidity (%)', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

# Step 4: Label x-axis and set title
ax1.set_xlabel('Date')
plt.title('Temperature and Humidity Over Time')

# Optional: Better layout and grid
fig.tight_layout()
plt.grid(True)
plt.show()
```

```python
'''Q.15 Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following
tasks using Matplotlib:
 a) Plot a histogram of the data with 30 bins.
 b) Overlay a line plot representing the normal distribution's probability density function (PDF).
 c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.
 d) Set the title of the plot as 'Histogram with PDF Overlay'.'''
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Step 1: Generate 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)  # Mean = 0, Std = 1

# Step 2: Plot the histogram
count, bins, ignored = plt.hist(data, bins=30, density=True, alpha=0.6, color='skyblue', edgecolor='black', label='Histogram')

# Step 3: Overlay PDF
x = np.linspace(min(data), max(data), 1000)
pdf = norm.pdf(x, loc=np.mean(data), scale=np.std(data))
plt.plot(x, pdf, 'r-', label='PDF', linewidth=2)

# Step 4: Label axes and title
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')
plt.title('Histogram with PDF Overlay')
plt.legend()

# Display the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
'''Q.17  Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the
origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.'''
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Generate random data
np.random.seed(42)
x = np.random.uniform(-10, 10, 200)
y = np.random.uniform(-10, 10, 200)

# Step 2: Determine quadrant for each point
def get_quadrant(x, y):
    if x > 0 and y > 0:
        return 'Quadrant I'
    elif x < 0 and y > 0:
        return 'Quadrant II'
    elif x < 0 and y < 0:
        return 'Quadrant III'
    elif x > 0 and y < 0:
        return 'Quadrant IV'
    else:
        return 'Origin/Axis'

# Apply the function to each point
quadrants = [get_quadrant(xi, yi) for xi, yi in zip(x, y)]

# Step 3: Create DataFrame
df = pd.DataFrame({'x': x, 'y': y, 'Quadrant': quadrants})

# Step 4: Create scatter plot with Seaborn
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='x', y='y', hue='Quadrant', palette='Set2', s=70)

# Step 5: Customize the plot
plt.axhline(0, color='gray', linestyle='--', linewidth=1)
plt.axvline(0, color='gray', linestyle='--', linewidth=1)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')
plt.legend(title='Quadrant')
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
'''Q.19 Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their
values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical
Bar Chart'''
```

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256
import numpy as np
import pandas as pd

# Use this if you're in Jupyter or Colab
output_notebook()

# Step 1: Generate random categorical data
categories = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
values = np.random.randint(10, 100, size=len(categories))

# Step 2: Create a DataFrame and data source
df = pd.DataFrame({'category': categories, 'value': values})
source = ColumnDataSource(df)

# Step 3: Create color mapper based on values
color_mapper = linear_cmap(field_name='value', palette=Viridis256, low=min(values), high=max(values))

# Step 4: Create Bokeh figure
p = figure(x_range=categories, height=400, width=600,
           title="Random Categorical Bar Chart",
           x_axis_label='Category', y_axis_label='Value',
           tools="pan,box_zoom,reset")

# Step 5: Draw bars with color and hover
p.vbar(x='category', top='value', width=0.6, source=source,
       fill_color=color_mapper, line_color='black', legend_label="Categories")

# Step 6: Add hover tool
hover = HoverTool(tooltips=[('Category', '@category'), ('Value', '@value')])
p.add_tools(hover)

# Customize appearance
p.legend.visible = False
p.grid.grid_line_alpha = 0.3

# Step 7: Show plot
show(p)
```
